```python
messages = [
    {
        "role": "system",
        "content": "You are a friendly chatbot who always responds in the style of a pirate",
    },
    {"role": "user", "content": "write me 5 points for AI forcast and its use cases"},
]
```

```python
start_resP = time.time()
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
outputs = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
print(outputs[0]["generated_text"])

end_resP = time.time()
time_resP = end_resP - start_resP
print("time_mL : ",time_resP)
```

# aihub-app/zyte-1B

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

model = AutoModelForCausalLM.from_pretrained("aihub-app/zyte-1B", torch_dtype=torch.bfloat16)
```

```python
tokenizer = AutoTokenizer.from_pretrained("aihub-app/zyte-1B")
```

```python
query = 'write me 5 points for AI forcast and its use cases'
prompt = f"<|system|> You are a helpful AI assistant.</s><|user|>{query}</s><|assistant|>"
```

```python
start_m1 = time.time()
inputs = tokenizer(prompt, return_tensors="pt", return_attention_mask=False)
outputs = model.generate(**inputs, max_length=256)
text = tokenizer.batch_decode(outputs)[0]
print(text)
end_m1 = time.time()
time_m1 = end_m1 - start_m1
print("time_mL : ",time_m1)
```
